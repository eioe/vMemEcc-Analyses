```python
from os import path as op
import os
import sys
import numpy as np
import pandas as pd
import autoreject
import mne
from library import config, helpers, preprocess
```

```python
# parse args:
# if (len(sys.argv) > 1):
#     helpers.print_msg('Running Job Nr. ' + sys.argv[1])
#     job_nr = int(sys.argv[1])
# else:
#     job_nr = None

%matplotlib qt

## Full procedure:
sub_list = np.setdiff1d(np.arange(10,config.n_subjects_total+1), config.ids_missing_subjects)
sub_list_str = ['VME_S%02d' % sub for sub in sub_list]

# if job_nr is not None:
#     sub_list_str = [sub_list_str[job_nr]]
for epo_part in ['stimon', 'cue', 'fulllength']: # []: #
    for subID in sub_list_str:
        fname = '-'.join([subID, epo_part, 'postICA'])
        try:
            data_pre = helpers.load_data(fname, config.paths['03_preproc-ica'], '-epo')
        except FileNotFoundError:
            print(f'No data for {subID}.')
            continue

        data_bl = data_pre.copy().apply_baseline((-config.times_dict['bl_dur_erp'], 0))

        data_post, ar, reject_log = preprocess.clean_with_ar_local(subID,
                                                                   data_bl,
                                                                   epo_part=epo_part,
                                                                   n_jobs=70,
                                                                   save_to_disc=True,
                                                                   ar_path=config.paths['03_preproc-ar'],
                                                                   rand_state=42)
```

```python
fpath = config.paths['03_preproc-ar']

for epo_part in ['cue', 'fulllength']: # 'stimon', 
    fpath_ar = op.join(config.paths['03_preproc-ar'], epo_part)
    files_ar = glob.glob(op.join(fpath_ar, f'*{epo_part}*-ar.fif'))
    files_ar.sort()
    fpath_csv = op.join(config.paths['03_preproc-ar'], epo_part, 'rejectlogs')
    files_csv = glob.glob(op.join(fpath_csv,f'*{epo_part}*.csv'))
    files_csv.sort()
    #print(files)

    for f_ar,f_csv in zip(files_ar, files_csv):
        rl_csv = pd.read_csv(f_csv)
        sID_csv = f_csv.split(op.sep)[-1].split('-')[0]
        sID_ar = f_ar.split(op.sep)[-1].split('-')[0]
        assert sID_ar == sID_csv
        sID = sID_csv
        # epo_part = f.split(op.sep)[-1].split('-')[1]
        ar = autoreject.read_auto_reject(f_ar)
        fpath_data = op.join(config.paths['03_preproc-ica'], 'cleaneddata', epo_part)
        data = helpers.load_data(f'{sID}-{epo_part}-postICA',
                                 path=fpath_data,
                                 append='-epo',
                                 verbose=False)
        data.apply_baseline((-0.2,0))
        data_c, rl = ar.transform(data, return_log=True)
        print(f'{sID}--{np.sum(rl_csv.badEpoch):<3}--{np.sum(rl.bad_epochs):<3}')
        # print(np.sum(rl.bad_epochs))
        fname_out = f'{sID}-{epo_part}-postAR'
        fpath_out = op.join(config.paths['03_preproc-ar'], epo_part, 'cleaneddata')
        helpers.save_data(data_c, fname_out, fpath_out, append='-epo')
```

```python

hp_evos_ga=dict()
hp_evos = dict()
hps = ['0.01', '0.05', '0.1', '0.5']
for hp in hps:
    fpath = op.join(config.paths['02_epochs'], hp, 'cue')
    files = os.listdir(fpath)
    evos = []
    for f in files:
        if op.isdir(op.join(fpath, f)): continue
        data = mne.read_epochs(op.join(fpath, f), verbose=False)
        evo = data.average()
        evos.append(evo)
    hp_evos_ga[hp] = mne.grand_average(evos)
    hp_evos[hp] = evos
```
